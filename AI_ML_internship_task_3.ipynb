```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn .metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
gh=fetch_california_housing(as_frame=True)
hg=gh.frame
print(hg.head())
```

```python
x=hg[['MedInc']]
y=hg['MedHouseVal']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
x=hg[['HouseAge','AveBedrms']]
y=hg[['MedHouseVal','AveOccup']]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
h=mean_absolute_error(y_test,y_pred)
g=mean_squared_error(y_test,y_pred)
v=r2_score(y_test,y_pred)
print("MAE:",h)
print("MSE:",g)
print("R²:",v)
```

```python
plt.scatter(x_test,y_test,color='blue',label='actual values')
plt.plot(x_test,y_pred,color='orange',label='regression lines')
plt.xlabel('HouseAge')
plt.ylabel('MedHouseVal')
plt.title("simple linear regression")
plt.legend()
plt.show()
```
